### Part 1

```python
def list_of_lists_to_array(list_of_lists):
    array = np.array([list(x) for x in list_of_lists])
    # replace all '.' with 0 and all '#' with 1
    start_y, start_x = np.where(array == 'S')
    array[array == '.'] = 0
    array[array == 'S'] = 0
    array[array == '#'] = 1
    array = array.astype(int)
    return array, start_x[0], start_y[0]

def get_neighbors(working_file_array, x, y):
    neighbours = list()
    # up
    if y > 0 and not working_file_array[y - 1, x]:
        neighbours.append((x, y - 1))
    # down
    if y < working_file_array.shape[0] - 1 and not working_file_array[y + 1, x]:
        neighbours.append((x, y + 1))
    # left
    if x > 0 and not working_file_array[y, x - 1]:
        neighbours.append((x - 1, y))
    # right
    if x < working_file_array.shape[1] - 1 and not working_file_array[y, x + 1]:
        neighbours.append((x + 1, y))
    return neighbours

def display(working_file, neighbors):
    for y in range(len(working_file)):
        for x in range(len(working_file[y])):
            if (x, y) in neighbors:
                print('O', end='')
            else:
                print(working_file[y][x], end='')
        print()
```

```python
working_file = file_content

working_file_array, start_x, start_y = list_of_lists_to_array(working_file)
neighbors = get_neighbors(working_file_array, start_x, start_y)
n_steps = 64
for i in tqdm(range(n_steps-1)):
    new_neighbors = list()
    for neighbor in neighbors:
        new_neighbors += get_neighbors(working_file_array, neighbor[0], neighbor[1])
        new_neighbors = list(set(new_neighbors))
    neighbors = new_neighbors
print(len(neighbors))
display(working_file, neighbors)
```
