### Part 1

```python

def translate_to_array(strings):
    # Get the length of the strings
    string_length = len(strings[0])

    # Create a 2D NumPy array with the correct shape
    array = np.empty((len(strings), string_length), dtype=int)

    # Iterate over the strings and replace each character with its position in the alphabet
    for i, string in enumerate(strings):
        for j, char in enumerate(string):
            array[i, j] = ord(char) - ord('a')

    return array
```

```python
with open('day12.txt', 'r') as f:
    lines = [x.replace('\n', '') for x in f.readlines()]

print(lines[:5])
grid = translate_to_array(lines)
start = np.where(grid == -14)
start_coordinates = [start[0][0], start[1][0]]
grid[start_coordinates[0], start_coordinates[1]] = 0

finish = np.where(grid == -28)
finish_coordinates = [finish[0][0], finish[1][0]]
grid[finish_coordinates[0], finish_coordinates[1]] = 25

print(start_coordinates)
print(finish_coordinates)
```

```python
def coord_to_str(square):
    return '({}, {})'.format(square[0], square[1])

def str_to_coord(str):
    return [int(str.split(', ')[0].replace('(', '')), int(str.split(', ')[1].replace(')', ''))]

to_visit = [coord_to_str(start_coordinates)]
visited = list()
i = 0
paths = list()
origins = dict()
while True:
    if i > 10000:
        break
    
    i += 1
    current_square = str_to_coord(to_visit.pop(0))
    visited.append(coord_to_str(current_square))

    # print(len(np.unique(visited)))
    # print(len(visited))
    # print(visited)
    # print()

    if current_square == finish_coordinates:
        print("Finished")
        break

    # check up
    if current_square[0] > 0:
        up = [current_square[0] - 1, current_square[1]]
        if coord_to_str(up) not in visited and coord_to_str(up) not in to_visit and grid[up[0], up[1]] - grid[current_square[0], current_square[1]] <= 1:
            to_visit.append(coord_to_str(up))
            origins['({}, {})'.format(up[0], up[1])] = '({}, {})'.format(current_square[0], current_square[1])
    # check right
    if current_square[1] < (len(grid[0]) - 1):
        right = [current_square[0], current_square[1] + 1]
        if coord_to_str(right) not in visited and coord_to_str(right) not in to_visit:
            if grid[right[0], right[1]] - grid[current_square[0], current_square[1]] <= 1:
                to_visit.append(coord_to_str(right))
                origins['({}, {})'.format(right[0], right[1])] = '({}, {})'.format(current_square[0], current_square[1])
    # check down
    if current_square[0] < (len(grid) - 1):
        down = [current_square[0] + 1, current_square[1]]
        if coord_to_str(down) not in visited and coord_to_str(down) not in to_visit and grid[down[0], down[1]] - grid[current_square[0], current_square[1]] <= 1:
            to_visit.append(coord_to_str(down))
            origins['({}, {})'.format(down[0], down[1])] = '({}, {})'.format(current_square[0], current_square[1])
    # check left
    if current_square[1] > 0:
        left = [current_square[0], current_square[1] - 1]
        if coord_to_str(left) not in visited and coord_to_str(left) not in to_visit and grid[left[0], left[1]] - grid[current_square[0], current_square[1]] <= 1:
            to_visit.append(coord_to_str(left))
            origins['({}, {})'.format(left[0], left[1])] = '({}, {})'.format(current_square[0], current_square[1])

    # print(to_visit)
origin = '({}, {})'.format(finish_coordinates[0], finish_coordinates[1])
track = list()
while origin != '({}, {})'.format(start_coordinates[0], start_coordinates[1]):
    track.insert(0, origin)
    origin = origins[origin]
print(track)
print(len(track))
```
