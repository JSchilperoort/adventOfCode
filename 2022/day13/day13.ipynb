### Part 1

```python
with open('day13.txt', 'r') as f:
    lines = [x.replace('\n', '') for x in f.readlines()]
lines = [eval(x) for x in lines if len(x) > 0 ]
print(lines[:5])
```

```python
def check_type(value):
    if isinstance(value, list):
        return 'list'
    elif isinstance(value, int):
        return 'integer'

def compare(left, right, outcome):
    for i in range(max(len(left), len(right))):
        if i >= len(left):
            # print('Left side ran out of items, so inputs are in the right order')
            return True
        elif i >= len(right):
            # print('Right side ran out of items, so inputs are not in the right order')
            return False

        if check_type(left[i]) == 'integer' and check_type(right[i]) == 'integer':
            if left[i] < right[i]:
                # print('Left side is smaller, so inputs are in the right order')
                return True
            elif right[i] < left[i]:
                # print('Right side is smaller, so inputs are not in the right order')
                return False
        elif check_type(left[i]) == 'list' and check_type(right[i]) == 'list':
            outcome = compare(left[i], right[i], outcome)
        
        else:
            if check_type(left[i]) == 'list':
                outcome = compare(left[i], [right[i]], outcome)
            else:
                outcome = compare([left[i]], right[i], outcome)
        if not outcome is None:
            return outcome

    return outcome

in_order_indices = list()
for i in range(0, len(lines), 2):
    
    left = lines[i]
    right = lines[i+1]
    if compare(left, right, None):
        in_order_indices.append(int((i/2)+1))
sum(in_order_indices)
```

```python
def check_type(value):
    if isinstance(value, list):
        return 'list'
    elif isinstance(value, int):
        return 'integer'

def compare(left, right, outcome):
    for i in range(max(len(left), len(right))):
        if i >= len(left):
            # print('Left side ran out of items, so inputs are in the right order')
            return 'Left'
        elif i >= len(right):
            # print('Right side ran out of items, so inputs are not in the right order')
            return 'Right'

        if check_type(left[i]) == 'integer' and check_type(right[i]) == 'integer':
            if left[i] < right[i]:
                # print('Left side is smaller, so inputs are in the right order')
                return 'Left'
            elif right[i] < left[i]:
                # print('Right side is smaller, so inputs are not in the right order')
                return 'Right'
        elif check_type(left[i]) == 'list' and check_type(right[i]) == 'list':
            outcome = compare(left[i], right[i], outcome)
        
        else:
            if check_type(left[i]) == 'list':
                outcome = compare(left[i], [right[i]], outcome)
            else:
                outcome = compare([left[i]], right[i], outcome)
        if not outcome is None:
            return outcome

    return outcome

in_order_indices = list()
for i in range(0, len(lines), 2):
    
    left = lines[i]
    right = lines[i+1]
    first_side = compare(left, right, None)
    print(first_side)
sum(in_order_indices)
```

```python
def sort_list(values):
    # Base case: the list is already sorted
    if len(values) <= 1:
        return values

    # Split the list into two halves
    mid = len(values) // 2
    left = values[:mid]
    right = values[mid:]

    # Recursively sort the two halves
    left = sort_list(left)
    right = sort_list(right)

    # Merge the sorted halves
    sorted_list = []
    while left and right:
        # Compare the first elements of the two lists
        if compare(left[0], right[0], None) == 'Left':
            # The left element is smaller, so add it to the sorted list
            sorted_list.append(left.pop(0))
        else:
            # The right element is smaller, so add it to the sorted list
            sorted_list.append(right.pop(0))

    # Add any remaining elements to the sorted list
    sorted_list.extend(left)
    sorted_list.extend(right)

    return sorted_list
divider_packets = [[[2]], [[6]]]
sorted = sort_list(lines + divider_packets)

print((sorted.index(divider_packets[0]) + 1) * (sorted.index(divider_packets[1]) + 1))
```
