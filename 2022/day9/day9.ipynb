```python
pos = [0, 0]
maxx = 0
maxy = 0
for instruction in input_data:
    direction, distance = instruction.split(' ')
    if direction == 'U':
        pos[0] += int(distance)
    elif direction == 'R':
        pos[1] += int(distance)
    elif direction == 'D':
        pos[0] -= int(distance)
    elif direction == 'L':
        pos[1] -= int(distance)
    if pos[0] > maxy:
        maxy = pos[0]
    if pos[1] > maxx:
        maxx = pos[1]

print(maxy)
print(maxx)
```

```python
def show_grid(maxy, maxx, pos_h, pos_t):
    for i in range(maxy + 1):
        for j in range(maxx + 1):
            if i == pos_h[0] and j == pos_h[1]:
                print('H', end='')
            elif i == pos_t[0] and j == pos_t[1]:
                print('T', end='')
            elif i == maxy and j == 0:
                print('s', end='')
            else:
                print('.', end='')
        print()
    print()
```
