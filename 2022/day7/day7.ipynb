```python
class FileOrDir:
    def __init__(self, name, parent):
        self.name = name
        self.parent = parent

class File(FileOrDir):
    def __init__(self, name, parent, size):
        super().__init__(name, parent)
        self.size = size
    
    def get_size(self):
        return self.size
    
    def __repr__(self):
        return "{} (file, size={})".format(self.name, self.get_size())

class Dir(FileOrDir):
    def __init__(self, name, parent):
        super().__init__(name, parent)
        self.children = list()

    def get_child(self, child_name):
        for child in self.children:
            if child.name == child_name:
                return child
        return None
    
    def set_child(self, child):
        self.children.append(child)
    
    def get_size(self):
        size = 0
        for child in self.children:
            size += child.get_size()
        return size
    
    def __repr__(self):
        return "{} (dir)".format(self.name)
```

```python
root_dir = Dir(name=lines[0].split(' ')[2], parent=None)
current_dir = root_dir
all_directories = [root_dir]
i = 1
while i < len(lines):
    if lines[i].startswith('$ '):
        instruction = lines[i].split(' ')[1]
        if instruction == 'cd':
            if lines[i].split(' ')[2] == '..':
                current_dir = current_dir.parent
            else:
                current_dir = current_dir.get_child(lines[i].split(' ')[2])
        elif instruction == 'ls':
            i += 1
            while i < len(lines) and not lines[i].startswith('$ '):
                if lines[i].startswith('dir '):
                    new_child = Dir(name=lines[i].split(' ')[1], parent=current_dir)
                    current_dir.set_child(new_child)
                    all_directories.append(new_child)
                else:
                    current_dir.set_child(File(name=lines[i].split(' ')[1], parent=current_dir, size=int(lines[i].split(' ')[0])))
                i += 1
            i -= 1
    i += 1

    

```
